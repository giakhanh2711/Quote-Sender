```python
with open("vietnamese.txt") as f:
    lines = f.readlines()

with open("vietnamese.txt", "w") as f:
    for line in lines:
        if line != "\n":
            lines.remove()
            line = line.split("“")[-1]
            line = '"' + line
            line = line.replace("”", '"')
            f.write(line)
```

```python
def find_closest_distance(point, segment):
    """
    Return np.array(2) landing point, and the distance
    """
    a, b = segment
    theta = np.linalg.norm(a - b) / ((a - b).T @ (point - b))

    candidates = segment
    distances = np.array([np.linalg.norm(point - endpoint) for endpoint in segment])

    if 0 <= theta <= 1:
        intersect = theta * a + (1 - theta) * b
        candidates = np.append(candidates, intersect)
        distances = np.append(distances, np.linalg.norm(point - intersect))
    
    optimal_point_idx = np.argmin(distances)

    return candidates[optimal_point_idx], distances[optimal_point_idx]
```
